```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import cv2 
import glob
import random
from os import listdir
import os
from tqdm import tqdm
```

```python
#getting all the images name for non IDC and IDC

root_dir = r'C:\Users\User\Desktop\UAB\2nd year\2nd semester\neural network-deep learning\project'
#dir_list = os.listdir(root_dir)

N_IDC = []
P_IDC = []

for dir_name in tqdm(os.listdir(root_dir)):
    dir_path = os.path.join(root_dir, dir_name)
    if os.path.isdir(dir_path):
        negative_dir_path = os.path.join(dir_path, '0')
        positive_dir_path = os.path.join(dir_path, '1')
        if os.path.isdir(negative_dir_path) and os.path.isdir(positive_dir_path):
            negative_image_paths = [
                os.path.join(negative_dir_path, image_name)
                for image_name in os.listdir(negative_dir_path)
                if image_name.endswith('.png')
            ]
            positive_image_paths = [
                os.path.join(positive_dir_path, image_name)
                for image_name in os.listdir(positive_dir_path)
                if image_name.endswith('.png')
            ]
            N_IDC.extend(negative_image_paths)
            P_IDC.extend(positive_image_paths)
```

```python
print (len(N_IDC))

print (len(P_IDC))
```

```python
import numpy as np
import keras.utils as image
import matplotlib.pyplot as plt

# Obtener 5 imágenes IDC negativas aleatorias
indices_negative = np.random.randint(0, len(N_IDC), 5)
plt.figure(figsize=(12, 6))
for i, idx in enumerate(indices_negative):
    img = image.load_img(N_IDC[idx], target_size=(100, 100))
    img = image.img_to_array(img)
    plt.subplot(2, 5, i+1)
    plt.title('IDC (-)')
    plt.imshow(img.astype('uint8'))
    plt.axis('off')

# Obtener 5 imágenes IDC positivas aleatorias
indices_positive = np.random.randint(0, len(P_IDC), 5)
for i, idx in enumerate(indices_positive):
    img2 = image.load_img(P_IDC[idx], target_size=(100, 100))
    img2 = image.img_to_array(img2)
    plt.subplot(2, 5, i+6)
    plt.title('IDC (+)')
    plt.imshow(img2.astype('uint8'))
    plt.axis('off')

plt.show()
```
